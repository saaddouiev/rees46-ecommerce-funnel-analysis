```python
import os
import numpy as np
import pandas as pd

file_path = "/Users/mac/Downloads/archive"

oct = pd.read_csv(os.path.join(file_path, "2019-Oct.csv" ))
nov = pd.read_csv(os.path.join(file_path, "2019-Nov.csv" ))
dec = pd.read_csv(os.path.join(file_path, "2019-Dec.csv" ))
jan = pd.read_csv(os.path.join(file_path, "2020-Jan.csv" ))
feb = pd.read_csv(os.path.join(file_path, "2020-Feb.csv" ))
```

```python
df = pd.concat([oct, nov, dec, jan, feb], ignore_index=True)

df.info()
```

```python
# Casting datatypes correctly
df['event_time'] = pd.to_datetime(df['event_time'], utc=True)
df['product_id'] = df['product_id'].astype('int32')
df['price'] = df['price'].astype('float32')
df['user_id'] = df['user_id'].astype('int32')
```

```python
df.isna().sum()
```

```python
# Handling nulls and missing values
df = df.drop(columns=['category_code'], errors='ignore')
df = df.dropna(subset=['user_session'])
df['brand'] = df['brand'].fillna('Unknown')
```

```python
# Eliminating duplicates
df.duplicated().sum()
df = df.drop_duplicates()
```
