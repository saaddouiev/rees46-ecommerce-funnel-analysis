```python
import pandas as pd
import plotly.express as px
import os


data = {
    'Stage': ['Product Views', 'Added to Cart', 'Purchased'],
    'Users': [1596359, 397390, 110518]
}

df = pd.DataFrame(data)

fig = px.funnel(
    df, 
    x='Users', 
    y='Stage', 
    title='User Journey: Lifetime Funnel Performance'
)

my_colors = ['#1F3A5F', '#2E6FA7', '#2FA88A']

fig.update_traces(
    marker=dict(color=my_colors),
    textinfo="value+percent initial", 
    textposition="inside"
)

fig.update_layout(
    yaxis_title="Funnel Stage",
    xaxis_title="Unique Users",
    font=dict(size=14)
)

```

```python
import plotly.graph_objects as go

fig = go.Figure(go.Bar(
    x=[8.2, 11],
    y=['Time in cart before abandonment', 'Full purchase journey (cart → purchase)'],
    orientation='h',
    marker_color=['#85B7EB', '#1F3A5F'],
    text=['8.2 days', '11.0 days'],
    textposition='outside',
    marker_line_width=0
))

fig.update_layout(
    title='Buying or abandoning: both take over a week',
    xaxis=dict(title='Average days', range=[0, 14], gridcolor='#e1e0d9'),
    yaxis=dict(automargin=True),
    plot_bgcolor='white',
    showlegend=False,
    margin=dict(l=260, t=60, b=50, r=60)
)

fig.add_annotation(
    x=0, y=-0.35, xref='x', yref='paper', xanchor='left',
    text="Session-level conversion is just 15.36% — most decisions unfold well after session one",
    showarrow=False, font=dict(size=12, color='gray')
)

fig.show()


```

```python

fig = go.Figure()


fig.add_trace(go.Bar(
    x=[1.44],
    y=['Abandoned Items'],
    orientation='h',
    name='Buys the same item later (1.44%)',
    marker_color='#1F3A5F',
    text=[''], 
    textposition='inside',
    marker_line_width=0
))

fig.add_trace(go.Bar(
    x=[30.0],
    y=['Abandoned Items'],
    orientation='h',
    name='Buys something else instead (30.0%)',
    marker_color='#85B7EB',
    text=['30.0%'],
    textposition='inside',
    textfont=dict(color='white', size=13),
    marker_line_width=0
))

fig.add_trace(go.Bar(
    x=[68.56],
    y=['Abandoned Items'],
    orientation='h',
    name='Never buys again (68.56%)',
    marker_color='#CBD5E1',
    text=['68.6%'],
    textposition='inside',
    textfont=dict(color='#1F3A5F', size=13),
    marker_line_width=0
))

fig.update_layout(
    title='What happens to an abandoned item: over two-thirds are lost forever',
    barmode='stack',
    xaxis=dict(
        title='Percentage of abandoned items', 
        range=[0, 100], 
        gridcolor='#e1e0d9', 
        ticksuffix='%'
    ),
    yaxis=dict(
        automargin=True, 
        showticklabels=False 
    ),
    plot_bgcolor='white',
    showlegend=True,
    legend=dict(
        orientation='h',
        yanchor='bottom',
        y=1.1,
        xanchor='left',
        x=0
    ),
    margin=dict(l=40, t=110, b=60, r=40)
)

fig.add_annotation(
    x=0, y=-0.35, xref='x', yref='paper', xanchor='left',
    text="Remarketing insight: Since only 1.44% buy the exact same item, focus your retargeting campaigns on cross-selling alternatives.",
    showarrow=False, font=dict(size=12, color='gray')
)

fig.show()

fig.write_html("what_happens_to_an_abandoned_item.html")
```

```python

fig = go.Figure()

# 1. The simple, clean timeline track
fig.add_trace(go.Scatter(
    x=[0, 14], y=[0, 0],
    mode="lines",
    line=dict(color="#E0E0E0", width=3),
    hoverinfo="skip"
))

# 2. The three core milestones
milestone_days = [0, 8.2, 11]
milestone_labels = ["<b>Day 0</b><br>Cart Add", "<b>Day 8.2</b><br>Avg. Abandonment", "<b>Day 11.0</b><br>Avg. Purchase"]

fig.add_trace(go.Scatter(
    x=milestone_days, y=[0, 0, 0],
    mode="markers+text",
    marker=dict(color=["#1F3A5F", "#D9534F", "#2FA88A"], size=14),
    text=milestone_labels,
    textposition="top center",
    textfont=dict(size=14, color="#333333")
))


fig.update_layout(
    title="<b>The 11-Day Decision Window:</b> Strategic & Technical Levers",
    showlegend=False,
    plot_bgcolor="#FFFFFF", 
    
    annotations=[
        
        dict(
            x=4.1, y=-0.1, 
            text="<b>1. The Marketing Lever (~8-Day Cart-Sit Window)</b><br>"
                 "Re-engagement should target the return, not the SKU.<br>"
                 "Trigger broad brand/category reminders to the 30% convertible pool.",
            showarrow=False,
            font=dict(size=12, color="#1F3A5F"),
            align="center"
        ),
       
        dict(
            x=11, y=-0.1, 
            text="<b>2. The Technical Lever (Cross-Session)</b><br>"
                 "Ensure strict cart persistence across devices.<br>"
                 "Friction here quietly discards warm demand.",
            showarrow=False,
            font=dict(size=12, color="#2FA88A"),
            align="center"
        )
    ],
    
   
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False, range=[-1, 14]),
    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False, range=[-0.25, 0.25]),
    height=350,
    margin=dict(t=60, b=20, l=20, r=20)
)

fig.show()

fig.write_html("11_day_decision_window.html")
```
